```python
#Let's check for NA and remove/recode where necessary
trainfeat.isnull().values.any()

#Looks like some have been recoded as negative values
trainfeat = trainfeat.where(trainfile>0,0)
```

```python
#Let's check for low variance columns
trainfeat.var().sort_values()[1:50]
#and columns that have few values
colnz = trainfeat.astype(bool).sum(axis=0)

len(colnz[colnz<50])

#some of these columns have very few (or zero) values, let's remove the lowest variance columns
var_threshold = .0005
lowvarcols = []

for col in trainfeat.columns:
    if trainfeat[col].var() <= var_threshold:
        lowvarcols.append(col)

print lowvarcols
len(lowvarcols)

trainfeat = trainfeat.drop(lowvarcols, axis=1)
```

```python
#Let's check for highly correlated columns
corr_thresh = .95
corrcols = []

corr_mat = trainfeat.corr().abs()

for i in range(corr_mat.shape[0]-1):
    for j in range(i+1, corr_mat.shape[1]):
        if corr_mat.values[i,j] >= corr_thresh and corr_mat.columns[j] not in corrcols:
            corrcols.append(corr_mat.columns[j])

print corrcols

trainfeat = trainfeat.drop(corrcols, axis=1)
```
