```python
import sys

import os
import numpy as np
import pandas as pd

from sklearn import cross_validation
sys.path.append('/home/ubuntu/xgboost/xgboost/python-package')
import xgboost as xgb
from sklearn.grid_search import GridSearchCV, RandomizedSearchCV
```

```python
trainfile = pd.read_pickle("./train_proc.pikl")
trainfileorig = pd.read_csv('./sant-train.csv')
trainlab = trainfileorig['TARGET']

testfile = pd.read_csv('./test_proc.csv')
testfile = testfile.drop(['ID','Unnamed: 0'], axis=1)
```

```python
X_train, X_test, y_train, y_test= cross_validation.train_test_split(trainfile, trainlab, test_size=0.25, random_state=777)
```

```python
eval_sets = [(X_train,y_train),(X_test,y_test)]
eval_metrics = ['logloss', 'auc']

#easy starting parameters, need a grid search next
xgbc = xgb.XGBClassifier(seed=777,learning_rate=0.05, n_estimators=50,min_child_weight=5,max_depth=4)
xgbc = xgbc.fit(X_train, y_train, eval_metric=eval_metrics, eval_set=eval_sets, verbose=True)

logloss = [xgbc.evals_result_['validation_%d' % i]['logloss'] for i in range(len(eval_sets))]
logloss = np.array(logloss, dtype=float).T
auc = [xgbc.evals_result_['validation_%d' % i]['auc'] for i in range(len(eval_sets))]
auc = np.array(auc, dtype=float).T

logloss_best_round = np.argmin(logloss, axis=0)
logloss_best = [logloss[logloss_best_round[0], 0], logloss[logloss_best_round[1], 1]]
auc_best_round = np.argmax(auc, axis=0)
auc_best = [auc[auc_best_round[0], 0], auc[auc_best_round[1], 1]]

print 'Best log loss train=%f (round=%d), test=%f (round=%d)' % (logloss_best[0], logloss_best_round[0], logloss_best[1], logloss_best_round[1])
print 'Best AUC train=%f (round=%d), test=%f (round=%d)' % (auc_best[0], auc_best_round[0], auc_best[1], auc_best_round[1])


plt.subplot(1, 2, 1)
plt.plot(logloss)
plt.legend(['Train', 'Test'])
plt.ylabel('Log loss')
plt.subplot(1, 2, 2)
plt.plot(auc)
plt.legend(['Train', 'Test'])
plt.ylabel('AUC')

xgb.plot_importance(xgbc)
```

```python
#easy starting parameters, need a grid search next

xgbc = xgb.XGBClassifier(seed=777, n_estimators=500,subsample=.5, learning_rate=0.02, max_depth=5, colsample_bytree=0.5)
parameters = {
        'reg_lambda': [1,1.1,1.2],
        'colsample_bytree':[.5,.55,.6,.65],
        'max_depth':[5,6]
}


#{'n_estimators': 500, 'subsample': 0.5, 'learning_rate': 0.02, 'colsample_bytree': 0.5, 'max_depth': 5}
#0.841231618306

# {'n_estimators': 400, 'subsample': 0.5, 'learning_rate': 0.02, 'colsample_bytree': 0.6, 'max_depth': 5}
# 0.841055350524
# parameters = {
#         'n_estimators':[10,20],
# }

n_iter_search = 24

xgb_cv = RandomizedSearchCV(xgbc, param_distributions=parameters,
                            n_iter=n_iter_search, scoring="roc_auc", n_jobs = 4, verbose=1)


xgbc = xgb_cv.fit(X_train, y_train)

print(xgbc.grid_scores_)
print(xgbc.best_params_)
print(xgbc.best_score_)
```
